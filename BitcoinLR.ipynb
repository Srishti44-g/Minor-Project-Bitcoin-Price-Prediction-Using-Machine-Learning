```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CleanedData.csv")
data.head()
```

```python
data.tail()
```

```python
x=data['Date'].values
y=data['Close'].values

plt.plot(x,y)
plt.title('Bitcoin price',fontsize=18)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Closing Price(USD)',fontsize=18)
plt.show()
```

```python
import datetime as dt
data['Date'] = pd.to_datetime(data['Date'])
data['Date']=data['Date'].map(dt.datetime.toordinal)

data.head()
```

```python
data.tail()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data.Date,data.Close)
```

```python
plt.scatter(X_train,Y_train,color='g',label='Training Data')
plt.scatter(X_test,Y_test,color='y',label='Testing Data')
plt.legend()
plt.title('Train Test Data')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X_train.values.reshape(-1,1),Y_train)

predict=reg.predict(X_test.values.reshape(-1,1))
```

```python
plt.plot(X_test,predict,label="Linear Regression")
plt.scatter(X_test,Y_test,color='#ef5423',label="Actual Test Data")
plt.xlabel("Date");
plt.ylabel("Closing Price(USD")
plt.title("Bitcoin Price Prediction Using LR")
plt.legend()
plt.show()
```

```python
reg.score(X_test.values.reshape(-1,1),Y_test.values)
```
